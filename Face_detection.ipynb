```python
#checking working of MTCNN
from mtcnn import MTCNN

detector = MTCNN()
print("MTCNN working ")
```

```python
input_video_path = r"C:\Users\sriii\OneDrive\Desktop\AI_Face_Detection_Project\input\test_video.mp4"
output_video_path = r"C:\Users\sriii\OneDrive\Desktop\AI_Face_Detection_Project\output\output_video.mp4"

cap = cv2.VideoCapture(input_video_path)

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

fps = cap.get(cv2.CAP_PROP_FPS)
if fps == 0:
    fps = 20

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

print("Processing with MTCNN...")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert BGR to RGB (IMPORTANT for MTCNN)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Detect faces
    faces = detector.detect_faces(rgb_frame)

    print("Faces detected:", len(faces))  # Debug

    for face in faces:
        x, y, w, h = face['box']
        
        # Draw rectangle
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)
        cv2.putText(frame, "Face", (x, y-10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0,255,0), 2)

    out.write(frame)

cap.release()
out.release()

print(" Done! Faces detected successfully.")
```

#### __Q4:__ Explain breifly model architectue of ResNet?
__ANS__  ResNet (Residual Network) is a deep learning model that uses skip connections to solve the vanishing gradient problem.

Instead of learning everything from scratch, it learns the difference (residual) between input and output. The input is directly added to the output of some layers.

This helps the model train very deep networks like 50+ layers easily and improves accuracy.
